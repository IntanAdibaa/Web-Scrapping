```python
from selenium import webdriver
import argparse
from webdriver_manager.chrome import ChromeDriverManager

def parse_args():
    parser = argparse.ArgumentParser()
    parser.add_argument('-i', '--infile', default='', help='input filename')
    parser.add_argument('-o', '--outfile', default='', help='output filename')
    return parser.parse_args()

def filter_text(text):
    lines = text.split('\n')
    filtered_lines = []
    for line in lines:
        if len(line.split()) >= 5:
            filtered_lines.append(line)
    return filtered_lines

def main():
    args = parse_args()
    outfile = args.outfile
    infile = args.infile

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--headless")

    with open(infile) as f:
        content = f.read().splitlines()

    full_text = []

    for u in content:
        driver = webdriver.Chrome( options=chrome_options)
        driver.get(u)
        elems = driver.find_element_by_tag_name('body').text
        filtered_text = filter_text(elems)
        full_text.append(filtered_text)
        driver.quit()
    
    print(full_text)

    with open(outfile, 'w') as f:
        for text_list in full_text:
            for line in text_list:
                f.write(line + '\n')
            f.write('\n')

if __name__ == "__main__":
    main()
```
